```python
df_encoded = pd.get_dummies(df, columns = ['type'], drop_first = True)
df_encoded
```

```python
X = df_encoded.drop(['quality'], axis = 1)
y = df_encoded.quality
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
model = DecisionTreeRegressor(random_state = 42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.3f}")
print(f"MSE: {mse:.3f}")
print(f"R²: {r2:.3f}")
```

DecisionTreeRegressor без настройки — переобучается.

Он запоминает тренировочные примеры, но плохо обобщает.

Почти никакой пользы при тестировании (R² ≈ 0).

```python
from sklearn.model_selection import GridSearchCV
```

```python
reg = DecisionTreeRegressor(random_state = 42)
param_grid = {
    'max_depth': [3,5,7,10,13,None],
    'min_samples_split': [2,5,10,15],
    'min_samples_leaf': [1,2,4,6]
}
```

```python
grid_search = GridSearchCV(reg,
                           param_grid = param_grid,
                           cv = 5,
                           scoring = 'neg_mean_squared_error',
                           n_jobs = -1,
                           verbose = 1)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("R²:", r2_score(y_test, y_pred))
```

### ✅ Результаты настройки DecisionTreeRegressor через GridSearchCV

Наилучшая модель была найдена при следующих параметрах:

- `max_depth = 5`
- `min_samples_split = 2`
- `min_samples_leaf = 6`

Оценка качества модели на тестовой выборке:

- **MAE (Mean Absolute Error)**: 0.571
- **MSE (Mean Squared Error)**: 0.554
- **R² (Коэффициент детерминации)**: 0.25

#### 🔍 Интерпретация

- **R² = 0.25** — это означает, что модель объясняет около 25% дисперсии в данных. Это **неплохой, но скромный результат**, который можно улучшить с помощью более мощных алгоритмов.
- DecisionTreeRegressor даёт интерпретируемую, но склонную к переобучению модель. Даже после настройки гиперпараметров его качество ограничено.


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfg = RandomForestRegressor(random_state = 42)
rfg.fit(X_train, y_train)
```

```python
y_pred_rf = rfg.predict(X_test)
mae = mean_absolute_error(y_test, y_pred_rf)
mse = mean_squared_error(y_test, y_pred_rf)
r2 = r2_score(y_test, y_pred_rf)

print(f"MAE: {mae:.3f}")
print(f"MSE: {mse:.3f}")
print(f"R²: {r2:.3f}")
```

```python
rf = RandomForestRegressor(random_state = 42)
grid_params_rfg = {
    'max_depth': [None, 4, 8, 11],
    'min_samples_split': [2,6,10],
    'n_estimators': [50, 100, 200],
    'max_features': ['log2', 'sqrt']
}
grid_search_rf = GridSearchCV(rf,
                              param_grid=grid_params_rfg,
                              cv = 5,
                              scoring = 'neg_mean_squared_error',
                              verbose = 2,
                              n_jobs = -1)
grid_search_rf.fit(X_train, y_train)
```

```python
grid_search_rf.best_params_
```

```python
best_model_rd = grid_search_rf.best_estimator_
y_pred_rf = best_model_rd.predict(X_test)
mae = mean_absolute_error(y_test, y_pred_rf)
mse = mean_squared_error(y_test, y_pred_rf)
r2 = r2_score(y_test, y_pred_rf)

print(f"MAE: {mae:.3f}")
print(f"MSE: {mse:.3f}")
print(f"R²: {r2:.3f}")
```

## Random Forest Regressor

На этом этапе мы применили модель `RandomForestRegressor` для решения задачи регрессии. Основные цели заключались в:

- улучшении качества предсказаний по сравнению с Decision Tree;
- подборе гиперпараметров с помощью `GridSearchCV`;
- интерпретации результатов модели.

 Модель показала улучшение по сравнению с базовым деревом решений. Особенно заметен рост коэффициента **детерминации (R²)**, что указывает на лучшее объяснение дисперсии целевой переменной.

**Вывод**

`Random Forest` позволил значительно улучшить результаты за счёт ансамблирования нескольких деревьев решений. Это делает модель более устойчивой к переобучению и позволяет лучше обобщать данные на тестовой выборке.





```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr = GradientBoostingRegressor(random_state = 42)

param_grid_gbr = {
    'n_estimators': [100, 200],
    'max_depth': [None, 2,5, 7],
    'subsample': [1.0, 0.8],
    'learning_rate': [0.1, 0.05]
}

grid_search_gb = GridSearchCV(gbr,
                              param_grid = param_grid_gbr,
                              cv = 5,
                              scoring = 'neg_mean_squared_error',
                              verbose = 2,
                              n_jobs = -1)

grid_search_gb.fit(X_train, y_train)
```

```python
grid_search_gb.best_params_
```

```python
best_model_gb = grid_search_gb.best_estimator_
y_pred_gbr = best_model_gb.predict(X_test)
print("MAE:", mean_absolute_error(y_test, y_pred_gbr))
print("MSE:", mean_squared_error(y_test, y_pred_gbr))
print("R²:", r2_score(y_test, y_pred_gbr))
```

Результаты на тестовой выборке:

MAE (средняя абсолютная ошибка): 0.409

MSE (среднеквадратичная ошибка): 0.361

R² (коэффициент детерминации): 0.510

Интерпретация:
R² ≈ 0.51 означает, что модель объясняет около 51% дисперсии целевой переменной. Это ощутимое улучшение по сравнению с базовым DecisionTreeRegressor и сравнимо с результатами RandomForestRegressor.

```python
from xgboost import XGBRegressor
```

```python
xgb = XGBRegressor(random_state = 42)
param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [None, 2, 4, 7],
    'subsample': [1.0, 0.8],
    'learning_rate': [0.1, 0.05],
}
grid_search_xgb = GridSearchCV(xgb,
                               param_grid = param_grid_xgb,
                               cv = 5,
                               scoring = 'neg_mean_squared_error',
                               n_jobs = -1,
                               verbose = 2)
grid_search_xgb.fit(X_train, y_train)
```

```python
grid_search_xgb.best_params_
```

```python
best_model_xgb = grid_search_xgb.best_estimator_
y_pred_xgb = best_model_xgb.predict(X_test)
print("MAE:", mean_absolute_error(y_test, y_pred_xgb))
print("MSE:", mean_squared_error(y_test, y_pred_xgb))
print("R²:", r2_score(y_test, y_pred_xgb))
```

```python
# Усреднённый блендинг
y_pred_blend = (y_pred_rf + y_pred_gbr + y_pred_xgb) / 3

print("Blended MAE:", mean_absolute_error(y_test, y_pred_blend))
print("Blended MSE:", mean_squared_error(y_test, y_pred_blend))
print("Blended R²:", r2_score(y_test, y_pred_blend))
```

📊 Вывод по модели блендинга
В рамках эксперимента был реализован усреднённый блендинг трёх моделей: `RandomForestRegressor, GradientBoostingRegressor и XGBRegressor`.
**Цель** — улучшить качество предсказания целевой переменной путём объединения сильных моделей


| Модель             | MAE       | MSE       | R²          |
| ------------------ | --------- | --------- | ----------- |
| RandomForest       | 0.428     | 0.353     | 0.522       |
| GradientBoosting   | 0.408     | 0.361     | 0.510       |
| XGBoost            | 0.44    | 0.37   | 0.49      |
| **Blending (Avg)** | **0.418** | **0.349** | **0.527** ✅ |

**Результат:**

Блендинг показал небольшое, но стабильное улучшение по всем метрикам **(MAE, MSE, R²)**, по сравнению с каждой из моделей по отдельности.
Это подтверждает идею, что ансамблирование нескольких сильных моделей может повысить **устойчивость **и **обобщающую способность** предсказаний.

```python
from sklearn.linear_model import Ridge
from sklearn.ensemble import StackingRegressor
```

```python
base_models = [
    ('rf', RandomForestRegressor(**grid_search_rf.best_params_, random_state=42)),
    ('gb', GradientBoostingRegressor(**grid_search_gb.best_params_, random_state = 42)),
    ('xgb', XGBRegressor(**grid_search_xgb.best_params_, random_state = 42))
]
# Мета-модель (на выходе)
meta_model = Ridge()

stack_model = StackingRegressor(
    estimators = base_models,
    final_estimator = meta_model,
    cv = 5,
    n_jobs = -1
)
stack_model.fit(X_train, y_train)
```

```python
y_pred_stack = stack_model.predict(X_test)
print("MAE:", mean_absolute_error(y_test, y_pred_stack))
print("MSE:", mean_squared_error(y_test, y_pred_stack))
print("R²:", r2_score(y_test, y_pred_stack))
```

В качестве базовых моделей использовались:

**RandomForestRegressor**

**GradientBoostingRegressor**

**XGBRegressor**

***Мета-модель: Ridge***

После применения стекинга модель показала следующие метрики:

`MAE: 0.413`

`MSE: 0.346`



`R²: 0.531`



📈 Это стало **лучшим** результатом по сравнению с одиночными моделями и блендингом.
